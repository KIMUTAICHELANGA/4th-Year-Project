```python
import pandas as pd

# Assuming the file was uploaded with the name 'kenya.csv'
df = pd.read_csv('kenya.csv')

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Check the column names
print(df.columns)
```

```python
import pandas as pd

# Assuming df is your DataFrame
columns_to_remove = ['Private equity(usd)', 'Pe number ', 'M&A(usd)', 'm & a amount']
df = df.drop(columns=columns_to_remove)

# If you want to confirm the columns have been removed
print(df.columns)
```

```python
df.head()
```

```python
# Check and print the data type of the 'Deal making' column
print(df['Deal making'].dtype)
```

```python
import pandas as pd

# Assuming your DataFrame is called df and the column is named 'Deal making'
# Replace commas with empty strings and then convert to numeric
df['Deal making'] = pd.to_numeric(df['Deal making'].str.replace(',', ''), errors='coerce')

# Print the DataFrame to check the result
print(df)
```

```python
import pandas as pd

# Assuming your DataFrame is called df
# Convert 'Deal making' column to numeric
df['Deal making'] = pd.to_numeric(df['Deal making'], errors='coerce')

# Impute NaN values with the mean of the column
mean_value = df['Deal making'].mean()
df['Deal making'].fillna(mean_value, inplace=True)

# Print the DataFrame to check the result
print(df)
```

```python
import pandas as pd

# Assuming your DataFrame is called df
# Assuming 'Country' and 'Year' are not to be considered for collinearity
columns_to_exclude = ['country', 'Year']

# Get the subset of columns to be considered for collinearity
columns_for_collinearity = df.drop(columns_to_exclude, axis=1)

# Calculate the correlation matrix
correlation_matrix = columns_for_collinearity.corr()

# Extract the correlation coefficients between 'Deal making' and other columns
deal_making_correlation = correlation_matrix['Deal making'].drop('Deal making')

# Print the correlation coefficients
print("Correlation coefficients between 'Deal making' and other columns:")
print(deal_making_correlation)
```

Based on the provided correlation coefficients, the columns that show no significant collinearity with the 'Deal making' column (correlation coefficients close to 0) are:

BroadMoneyGrowth
GrossNationalSavings
ImportsOfGoodsAndServices(GDP)
NetForeignDirectInvestment(GDP)
PerCapitaGDPGrowth
TermsOfTrade
TotalInvestment
TradeBalance(GDP)
These columns have correlation coefficients close to 0 (around -0.1 to 0.1), indicating little linear relationship with the 'Deal making' column. Therefore, they can be considered as not showing significant collinearity with the 'Deal making' column.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
correlation_matrix = df.drop(['country', 'Year'], axis=1).corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# List of columns with correlation coefficients close to 0
columns_to_remove = ['BroadMoneyGrowth', 'GrossNationalSavings', 'ImportsOfGoodsAndServices(GDP)',
                     'NetForeignDirectInvestment(GDP)', 'PerCapitaGDPGrowth', 'TermsOfTrade',
                     'TotalInvestment', 'TradeBalance(GDP)']

# Drop the columns from the DataFrame
df_filtered = df.drop(columns_to_remove, axis=1)

# Print the DataFrame after removing the columns
print(df_filtered.head())
```

```python
# Save the new DataFrame to a new CSV file
df_filtered.to_csv('newkenya.csv', index=False)

# Confirm the file has been saved
print("New CSV file 'newkenya.csv' has been saved with the removed columns.")
```
